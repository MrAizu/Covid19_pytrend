```python
# Connecting To A Google

import pandas as pd
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=['Covid 19 Prevention'])
#  interest by region
data =pytrend.interest_by_region()
data.head(10)
```

```python
pd.set_option('display.max_rows',None,'display.max_columns',None)
data
```

```python
data[['Covid 19 Prevention']].idxmax()
```

```python
# Plot The Analysis Result on Bar Graph
data.reset_index().plot(x='geoName', y='Covid 19 Prevention',figsize=(120,10),
                        kind='bar')
```

```python
related_queries = pytrend.related_queries()
related_queries.values()
```

```python
# Search Covid 19 prevention search term in Malaysia
kw_list= ['covid 19 prevention','coronavirus prevention','covid-19 prevention',
          'prevention of covid 19']
pytrend.build_payload(kw_list, timeframe='2019-04-28 2021-05-17', geo='MY')


# timeframe ----> YYYY-MM-DD
# geo ---> Two Letters Country Abbreviation
```

```python
malaysia_data = pytrend.interest_over_time()
malaysia_data 
```

```python
import matplotlib.pyplot as plt
data = malaysia_data.drop(labels='isPartial',axis='columns')
image = data.plot(title='Covid 19 Prevention Search Term Interest In Malaysia')
fig = image.get_figure()
fig.set_size_inches(18.5,10.5)
```

```python
# Saving analysis result in csv file
data.to_csv('search_trend.csv')
```
